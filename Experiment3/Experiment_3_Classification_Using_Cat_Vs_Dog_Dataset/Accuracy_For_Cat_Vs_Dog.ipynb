```python
import re
```

```python
# Load the Log File
log_file_path = "/kaggle/input/cat-vs-dog-logs/download.txt"
with open(log_file_path, "r") as file:
    log_data = file.readlines()
```

```python
# Define Regex Batterns to Extract Validation Loss and Accuracy
val_loss_pattern = re.compile(r"Validation Loss: ([\d.]+)")
val_acc_pattern = re.compile(r"Validation Accuracy: ([\d.]+)%")
```

```python
# Variables to Store the Last Recorded Values
final_val_loss = None
final_val_acc = None
```

```python
# Iterate Through the Log File Lines
for line in log_data:
    val_loss_match = val_loss_pattern.search(line)
    val_acc_match = val_acc_pattern.search(line)

    if val_loss_match and val_acc_match:
        final_val_loss = float(val_loss_match.group(1))
        final_val_acc = float(val_acc_match.group(1))
```

```python
# Print Final Results
print(f"Final Test Loss: {final_val_loss}")
print(f"Final Total Accuracy: {final_val_acc}%")
```
